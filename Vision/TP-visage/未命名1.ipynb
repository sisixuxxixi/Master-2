```python
def Bbox(file):
    f = open(file.split(".")[0] + ".pts")
    landmarks = []
    min_x = np.inf
    max_x =-1
    min_y = np.inf
    max_y =-1
    landmarks=[]
    for line in f.readlines():
        x_ldm = float(line.split(" ")[0])
        y_ldm = float(line.split(" ")[1].split("\n")[0])
        landmarks.append([x_ldm, y_ldm])
        if x_ldm < min_x:
            min_x=x_ldm  
        if x_ldm >= max_x:
            max_x = x_ldm
        if y_ldm < min_y:
            min_y=y_ldm 
        if y_ldm >= max_y:
            max_y = y_ldm
    return [min_x, max_x, min_y, max_y], landmarks
    
```
