```python
from ultralytics import YOLO
import torch
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Image
import matplotlib.patches as patches
import os
```

```python
Model = YOLO("yolov8x.pt")
```

```python
if torch.cuda.is_available():
    Model = Model.to('cuda')
```

```python
Model.train(data='data_resolution.yaml', epochs=300, batch = 32, imgsz = 1280, patience=300, optimizer = 'AdamW', lr0 = 0.00000001)
```

```python
test_set = ['2ec3f1bb9','3589adb90','d488c759a','aa05346ff','57512b7f1']
path = 'dataset_resolution/images/test/'
```

```python
model_test = YOLO("runs/detect/train2/weights/best.pt")
for image_id in test_set:
    rest_path = os.path.join(path, f"{image_id}.png")
    results = model_test.predict(rest_path)
    
    for result in results:
        orig_img = result.orig_img
        boxes = result.boxes
        
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    res_plotted = results[0].plot()
    axs[0].imshow(res_plotted)
    axs[0].set_title('Result Plot')


    axs[1].imshow(orig_img) 
    axs[1].set_title('Boxes Plot')


    for i, box in enumerate(boxes.xyxy):
        x1 = box[0].item()
        y1 = box[1].item()
        x2 = box[2].item()
        y2 = box[3].item()
        width, height = x2 - x1, y2 - y1
        rect = patches.Rectangle((x1, y1), width, height, linewidth=1, edgecolor='r', facecolor='none')
        axs[1].add_patch(rect)

    plt.tight_layout()  
    plt.show()
```
